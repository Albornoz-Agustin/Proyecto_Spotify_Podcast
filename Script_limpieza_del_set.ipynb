```python
# Convertir idioma_show a mayúsculas #
df_fechas_2019_2024['idioma_show'] = df_fechas_2019_2024['idioma_show'].str.upper()
```

```python
# Renombrar columna 'keyword' a 'categoria' #
df_fechas_2019_2024.rename(columns={'keyword': 'categoria'}, inplace=True)
```

```python
# Traducir las categorías del inglés al español #

traducciones_categoria = {
    'technology': 'tecnología',
    'news': 'noticias',
    'comedy': 'comedia',
    'sports': 'deportes',
}

df_fechas_2019_2024['categoria'] = df_fechas_2019_2024['categoria'].replace(traducciones_categoria)

# Capitalizar la primera letra de cada categoría #
df_fechas_2019_2024['categoria'] = df_fechas_2019_2024['categoria'].str.capitalize()
```

```python
# Limpiar guiones bajos de 'nombre_podcast' y 'editorial' #

df_fechas_2019_2024['nombre_podcast'] = df_fechas_2019_2024['nombre_podcast'].str.replace('_', ' ', regex=False)
df_fechas_2019_2024['editorial'] = df_fechas_2019_2024['editorial'].str.replace('_', ' ', regex=False)
```

```python
# Incorporamos la columna de proporcion de episodios descargados para tener una metrica de medicion #

df_fechas_2019_2024['proporcion_episodios'] = (
    df_fechas_2019_2024['episodios_descargados'] / df_fechas_2019_2024['total_episodios']
).round(2)
```

```python
# Reorganizar columnas: mover 'proporcion_episodios' después de 'episodios_descargados' #

cols = list(df_fechas_2019_2024.columns)
cols.remove('proporcion_episodios')
pos = cols.index('episodios_descargados') + 1
cols.insert(pos, 'proporcion_episodios')

# Reordenar el DataFrame #
df_fechas_2019_2024 = df_fechas_2019_2024[cols]
```

```python
# Limpieza de datos en las columnas de caracteres no necesarios #

df_fechas_2019_2024['nombre_episodio'] = df_fechas_2019_2024['nombre_episodio'].str.lower()

df_fechas_2019_2024['nombre_episodio'] = df_fechas_2019_2024['nombre_episodio'].apply(
    lambda x: re.sub(r'[^\w\s]', '', x)
)

palabras_eliminar = ['episodio', 'capítulo', 'cap', 'ep', 'podcast']
for palabra in palabras_eliminar:
    df_fechas_2019_2024['nombre_episodio'] = df_fechas_2019_2024['nombre_episodio'].str.replace(palabra, '', regex=False)

def eliminar_nombre_podcast(ep, podcast):
    return ep.replace(podcast.lower(), '').strip()

df_fechas_2019_2024['nombre_episodio'] = df_fechas_2019_2024.apply(
    lambda row: eliminar_nombre_podcast(row['nombre_episodio'], row['nombre_podcast']), axis=1
)
```

```python
# Función para clasificar los episodios por duración #

def categorizar_duracion(minutos):
    if minutos < 10:
        return 'muy corto'
    elif 10 <= minutos < 20:
        return 'corto'
    elif 20 <= minutos <= 45:
        return 'medio'
    elif 45 < minutos <= 60:
        return 'largo'
    else:
        return 'muy largo'

# Crear columna 'categoria_duracion' #
df_fechas_2019_2024['categoria_duracion'] = df_fechas_2019_2024['duracion_min'].apply(categorizar_duracion).astype('category')
```

```python
# Función para categorizar cada episodio según la fecha de lanzamiento #

def categorizar_periodo(fecha):
    if fecha < pd.Timestamp('2020-03-01'):
        return 'pre-pandemia'
    elif pd.Timestamp('2020-03-01') <= fecha <= pd.Timestamp('2021-12-31'):
        return 'pandemia'
    else:
        return 'post-pandemia'

# Crear nueva columna 'periodo_pandemia' #
df_fechas_2019_2024['periodo_pandemia'] = df_fechas_2019_2024['fecha_lanzamiento'].apply(categorizar_periodo).astype('category')
```

```python
# Se incorporan mes y año #

df_fechas_2019_2024['año'] = df_fechas_2019_2024['fecha_lanzamiento'].dt.year
df_fechas_2019_2024['mes'] = df_fechas_2019_2024['fecha_lanzamiento'].dt.month

cols = list(df_fechas_2019_2024.columns)
for col in ['año', 'mes']:
    if col in cols:
        cols.remove(col)

pos = cols.index('fecha_lanzamiento') + 1
cols[pos:pos] = ['año', 'mes']   # inserta año y mes en bloque

df_fechas_2019_2024 = df_fechas_2019_2024[cols]
```
